```python
import re
from urllib.parse import urlparse

def extract_features(url, feature_names):
    features = []
    parsed = urlparse(url)
    hostname = parsed.netloc

    # 1. having_IP_Address
    features.append(-1 if re.search(r'\d+\.\d+\.\d+\.\d+', hostname) else 1)

    # 2. URL_Length
    if len(url) < 54:
        features.append(1)
    elif 54 <= len(url) <= 75:
        features.append(0)
    else:
        features.append(-1)

    # 3. Shortining_Service
    features.append(-1 if any(short in url for short in ["bit.ly","tinyurl","goo.gl"]) else 1)

    # 4. having_At_Symbol
    features.append(-1 if "@" in url else 1)

    # 5. double_slash_redirecting
    features.append(-1 if "//" in url[7:] else 1)

    # 6. Prefix_Suffix
    features.append(-1 if "-" in hostname else 1)

    # 7. having_Sub_Domain
    features.append(-1 if hostname.count('.') > 1 else 1)

    # 8. SSLfinal_State
    features.append(1 if parsed.scheme == "https" else -1)

    # Fill remaining features with neutral value (0)
    while len(features) < len(feature_names):
        features.append(0)

    return features
```

```python
url = "tinyurl.com/bswqloj"

features = extract_features(url, feature_names)
sample = pd.DataFrame([features], columns=feature_names)

prediction = dt.predict(sample)[0]
prob = dt.predict_proba(sample)

print("Prediction:", "Phishing" if prediction == 1 else "Legitimate")
print("Confidence:", prob)
```
